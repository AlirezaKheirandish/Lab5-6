## OOP

```python
class Set:
    def __init__(self):
        self.collection = list()

    def __and__(self, other):
        return [x for x in self.collection if x in other.collection]

    def __or__(self, other):
        Join = self.collection + other.collection
        Cleanset = list()
        [Cleanset.append(x) for x in Join if x not in Cleanset]
        return Cleanset

    def __eq__(self, other):
        return self.collection == other.collection

    def __str__(self):
        return f"{self.collection}"

    def add(self, value):
        try:
            if len(value) > 0:
                pass
        except:
            value = [value]
        self.collection += value
        Cleanset = list()
        [Cleanset.append(x) for x in self.collection if x not in Cleanset]
        self.collection = Cleanset

    def delete(self, value):
        try:
            if len(value) > 0:
                pass
        except:
            value = [value]
        [self.collection.remove(x) for x in value if x in self.collection]
```

```python
A = Set()
B = Set()
C = Set()
print(f"A={A}\nB={B}\nC={C}")

A.add([1, 3, 4, 5, 6, 7, 8, 9])
B.add([1, 2, 10])
C.add(1)
C.add(4)
C.add(5)
C.add(6)
print(f"A={A}\nB={B}\nC={C}")
```

```python
A.delete(10)
B.delete(5)
print(f"Delete:\nA={A}\nB={B}")

Join = A | B
InterSection = A & B
print(f"Union-Intersection Principle:\nA | B = {Join}\nA & B = {InterSection}")
Equal_1 = A == B
D = Set()
D.add([1, 4, 5, 6])
Equal_2 = C == D
print(f"Equality:\nA == B -> {Equal_1}")
print(f"C == D -> {Equal_2}")
```

## Numpy & Scipy

```python
import numpy as np
from scipy.fft import fft, fftshift
from scipy import linalg
```

```python
initial_array = np.arange(start=1, stop=7, step=1)
reshaped_array = np.reshape(initial_array, (3, 2))
print(reshaped_array)
another_array = np.random.randint(low=1, high=20, size=(3, 3))
print(another_array)
stacked_array = np.hstack([reshaped_array, another_array])
print(stacked_array)
flat_array = np.ravel(stacked_array)
print(flat_array)

max_element_ind = np.argmax(flat_array)
print(f"maximum value index in flat_array = {max_element_ind} with value = {flat_array[max_element_ind]}")
```

```python
A = np.mat([[3, 1, 4], [1, 5, 9], [2, 6, 5]])
b = np.mat([2, 4, -1])
A_tran = A.transpose()
A_inv = linalg.inv(A)
A_det = linalg.det(A)

print(f"transpose of matrix A =\n {A_tran}\n")
print(f"inverse of matrix A =\n {A_inv}\n")
print(f"Determinant of matrix A =\n {A_det}\n")
```

```python
A_eigvals, A_eigvects = linalg.eig(A)
U, sigma, V_star = linalg.svd(A)
Q, R = linalg.qr(A)


print(f"Eigenvalues of matrix A =\n{A_eigvals}\nEigenvectors of matrix A =\n{A_eigvects}\n")
print(f"U matrix of SVD(A) =\n{U}\nsigma matrix of SVD(A) =\n{sigma}\nV* matrix of SVD(A) =\n{V_star}\n")
print(f"Q matrix of QR(A) =\n{Q}\nR matrix of QR(A) =\n{R}\n")
```

```python
X = linalg.solve(A, b.T)
print(f"X for equation AX = bT:\n{X}")
```

```python
t = np.random.randint(low=-100, high=100, size=(1, 1000))
t_p = np.interp(t, (t.min(), t.max()), (-10*np.pi, +10*np.pi))
t_sorted = np.sort(t_p)
t_sorted = t_sorted[0]
plt.plot((t_sorted))
plt.show()
```

```python
y = np.sinc(t_sorted)
_ = plt.plot(t_sorted,y)
```

```python
z = fft(y)
z = np.abs(fftshift(z))
```

```python
fig, axs = plt.subplots(2, figsize=(10, 10))

axs[0].plot(t_sorted, y)
axs[0].set_title("sinc function")
axs[0].set_ylabel("sinc value")
axs[0].set_xlabel("t")
axs[0].grid()
axs[0].plot()

axs[1].plot(z)
axs[1].set_title("fft(sinc) function")
axs[1].set_ylabel("fft value")
```
